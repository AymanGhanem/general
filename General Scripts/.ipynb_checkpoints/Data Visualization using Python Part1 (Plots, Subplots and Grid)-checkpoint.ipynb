```python
#Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Basic visualizations -- One plot figure
x = np.linspace(-5,5,20)
plt.figure(figsize = (20,8))
plt.title("First Plot",fontsize = 24)
y1 = x**2
y2 = -x**2
plt.plot(x,y1,label = "y = x**2",linewidth = 6,color = 'red',marker = "o",markersize = 16)
plt.plot(x,y2,label = "y = -x**2",linewidth = 6,color = 'blue',marker = "*",markersize = 16)
plt.plot([0,0,0],[-25,0,25],label = "Vertical Line",linewidth = 8,color = 'black')
plt.legend(fontsize = 18,title = 'Equations')
plt.xlabel("x",fontsize = 18)
plt.ylabel("y-axis",fontsize = 18)
plt.xticks(range(-5,6,1))
plt.yticks(range(-25,30,5))
plt.show()
```

```python
#Subplots
plt.figure(figsize=(20,8))

#subplot parameters : rows, columns, panel selected
plt.subplot(1, 2, 1)
plt.title("First Subplot")
plt.plot(x,y1,label = "y = x**2",linewidth = 6,color = 'orange')
plt.plot([0,0,0],[-25,0,25],label = "Vertical Line",linewidth = 8,color = 'red')
plt.legend(fontsize = 18,title = 'Equations')
plt.xlabel("x",fontsize = 18)
plt.ylabel("y-axis",fontsize = 18)
plt.xticks(range(-5,6,1))
plt.yticks(range(-25,30,5))

plt.subplot(1, 2, 2)
plt.title("Second Subplot")
plt.plot(x,y2,label = "y = -x**2",linewidth = 6,color = 'yellow')
plt.plot([0,0,0],[-25,0,25],label = "Vertical Line",linewidth = 8,color = 'red')
plt.legend(fontsize = 18,title = 'Equations')
plt.xlabel("x",fontsize = 18)
plt.ylabel("y-axis",fontsize = 18)
plt.xticks(range(-5,6,1))
plt.yticks(range(-25,30,5))

plt.show()
```

```python
#Using another layer of matplotlib
figue , axes = plt.subplots(figsize = (20,8))

axes.plot(
    x ,
    y1 ,
    linewidth = 6 ,
    marker = 'o' ,
    markersize = 20 ,
    color = 'green' ,
    label = 'y = x**2' ,
)
axes.plot(
    x ,
    y2 ,
    linewidth = 6 ,
    marker = 'P' ,
    markersize = 20 ,
    color = 'blue' ,
    label = 'y = -x**2' ,
)
axes.plot(
    [0,0,0] ,
    [-25,0,25] ,
    linewidth = 6 ,
    marker = 'X' ,
    markersize = 20 ,
    color = 'red' ,
    label = 'Vertical Line' ,
    linestyle = 'dotted'
)

axes.set_title("New Plot without using artist layer!",fontsize= 22)
axes.set_xlabel("X",fontsize = 24)
axes.set_ylabel("Y",fontsize = 24)
axes.legend(fontsize = 22)

plt.show()
```

```python
#Markers and Linestyles
markers = [marker for marker in plt.Line2D.markers]
linestyles = ['-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted']
print('Markers: {}'.format(markers))
print("----------------------------------------------------------------------------")
print('Line styles: {}'.format(linestyles))
```

```python
#Line Styles
figue , axes = plt.subplots(figsize = (20,8))

axes.plot(
    x ,
    x ,
    linewidth = 2 ,
    marker = 'v' ,
    markersize = 6 ,
    color = 'green' ,
    label = 'dotted' ,
    linestyle = ":"
)
axes.plot(
    x ,
    x + 1 ,
    linewidth = 2 ,
    marker = 'v' ,
    markersize = 6 ,
    color = 'red' ,
    label = 'dashed' ,
    linestyle = "--"
)
axes.plot(
    x ,
    x + 2 ,
    linewidth = 2 ,
    marker = 'v' ,
    markersize = 6 ,
    color = 'orange' ,
    label = 'solid' ,
    linestyle = "-"
)
axes.plot(
    x ,
    x + 3 ,
    linewidth = 2 ,
    marker = 'v' ,
    markersize = 6 ,
    color = 'gray' ,
    label = 'None' ,
    linestyle = "None"
)
axes.plot(
    x ,
    x + 4 ,
    linewidth = 2 ,
    marker = 'v' ,
    markersize = 6 ,
    color = 'black' ,
    label = 'dashdot' ,
    linestyle = "-."
)

axes.set_title("Line Styles",fontsize = 26)
axes.legend(fontsize = 22)

plt.show()
```

```python
#Markers
figue , axes = plt.subplots(figsize = (20,33))

for index , marker in enumerate(markers) :
    axes.plot(
        x ,
        x + index ,
        linewidth = 2 ,
        marker = marker ,
        markersize = 6 ,
        label = marker
    )
axes.legend(fontsize = 12)    
plt.show()
```

```python
#Figures and subfigures
figure , axes = plt.subplots(figsize = (20,8))

axes.plot(
    range(50) ,
    np.random.normal(1,1,50) ,
    linewidth = 2 ,
    marker = '+' ,
    markersize = 10 ,
    color = 'blue' ,
)

axes.set_title("Figure",fontsize = 26)
plt.show()
```

```python
#Figures and subfigures
figure , ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows = 2,ncols = 2,figsize = (20,8))

ax0.plot(np.random.normal(1,2,50), c='black')
ax1.plot(np.random.normal(0.4,2,50), c='red', linestyle='--')
ax2.plot(np.random.normal(1,1,50), c='orange', linestyle=':',linewidth = 3)
ax3.plot(np.random.normal(-1,2,50), c='blue', marker='o', linewidth=3.0)

ax0.set_title("Sub-figure",fontsize = 12)
ax1.set_title("Sub-figure",fontsize = 12)
ax2.set_title("Sub-figure",fontsize = 12)
ax3.set_title("Sub-figure",fontsize = 12)

plt.show()
```

```python
#Subplots with subplot2grid
plt.figure(figsize = (20,8))

ax0 = plt.subplot2grid((3,3),(0,0),colspan = 2)
ax0.plot(np.random.normal(1,2,50), c='black')
ax0.set_title("ax0")
ax0.grid(True)
ax0.axis(True)
ax0.set_xticks([0,49])

ax1 = plt.subplot2grid((3,3),(0,2),colspan = 2)
ax1.plot(np.random.normal(1,9,50), c='red')
ax1.set_title("ax1")

ax2 = plt.subplot2grid((3,3),(1,1),rowspan = 2,colspan = 2)
ax2.plot(np.random.normal(1,2,50), c='orange',linewidth = 4,linestyle = 'dotted')
ax2.set_title("ax2")

ax3 = plt.subplot2grid((3,3),(1,0))
ax3.plot(np.random.normal(1,2,50), c='blue',linewidth = 4,linestyle = 'dashed')
ax3.set_title("ax3")

ax4 = plt.subplot2grid((3,3),(2,0))
ax4.plot(np.random.normal(1,2,50), c='green',linewidth = 4,linestyle = 'solid')
ax4.set_title("ax4")

plt.show()
```
