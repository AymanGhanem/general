<h1 style = "letter-spacing : 5px ; font-weight : bold" class = "text-primary">Simple Example of Word Cloud</h1>

Here we'll visualize the most repeated words of a text in a good looking way.
Just that ^-^  .

Importing all neccessary packages

```python
import os
import re
import string
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud , STOPWORDS
from PIL import Image
```

Importing datasets

```python
base_path = os.getcwd()
alice_path = os.path.join(base_path,"Files\\alice_novel.txt")
alice_novel = open(alice_path,'r').read()
print(alice_novel[1050:1300])
```

# Cleaning dataset
<p>Doing just the simplest operations to process the text</p>
<p class = "bg-warning text-danger">We're not going into NLP.</p>

```python
#To ignore the influence of characters' cases.
alice_novel = alice_novel.upper()
print(alice_novel[1050:1300])
```

```python
#Words that mean nothing valuable and don't add any value to the summary.
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
#Updating the stopwords
new_stop_words = ["must","oh","two","said",'now']
for word in new_stop_words :
    stopwords.add(word)
```

# Visualizing the word  cloud

<h3>Word cloud</h3>
<i>Word clouds</i> work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# Instantiate a word cloud object
alice_wc = WordCloud(
    background_color = 'black',
    max_words = 2000,
    stopwords = stopwords
)

# Generate the word cloud
alice_wc.generate(alice_novel)
print("The generated object " ,alice_wc)
print("The type of the generated object " , type(alice_wc))
```

```python
# display the word cloud
plt.figure(figsize = (20,8))
plt.axis("off")
plt.imshow(alice_wc, interpolation='bilinear')
plt.show()
```

```python
mask_path = os.path.join(base_path,"Images\\alice_mask.png")
alice_mask = np.array(Image.open(mask_path))
plt.figure(figsize = (20,8))
plt.axis("off")
plt.imshow(alice_mask,cmap = plt.cm.gray)
plt.show()
```

```python
# Instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)

# Generate the word cloud
alice_wc.generate(alice_novel)

# Define the dimensions of a figure
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(8)

# Display the word cloud
plt.imshow(alice_wc,cmap = plt.cm.gray)
plt.axis('off')
plt.show()
```
