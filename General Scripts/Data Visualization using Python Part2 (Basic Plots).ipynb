```python
#Import needed packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#Scatter plot
x = np.random.randn(20)
y = np.random.normal(-1,2,20)

plt.figure(figsize = (20,8))
plt.scatter(x,y,c=y,s=(y + abs(min(y)))*100,cmap = 'gist_heat',alpha = 0.6)
plt.colorbar()
plt.title("Scatter Plot",fontsize = 28)
plt.xlabel("X",fontsize = 20)
plt.ylabel("Y",fontsize = 20)

plt.show()
```

```python
#Scatter plot with subplots
x = np.random.randn(20)
y = np.random.normal(-1,2,20)

fig = plt.figure(figsize = (20,8))
ax1 = fig.add_subplot(1,2,1)
plt.scatter(x,y,c=y,s=(y + abs(min(y)))*100,cmap = 'PiYG',alpha = 0.8)
plt.colorbar()
plt.title("Scatter Plot",fontsize = 28)
plt.xlabel("X",fontsize = 20)
plt.ylabel("Y",fontsize = 20)

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x,y,c=y,s=(y + abs(min(y)))*100,cmap = 'rainbow',alpha = 0.8,edgecolor = 'black',linewidth = 7)
plt.colorbar()
plt.title("Scatter Plot",fontsize = 28)
plt.xlabel("X",fontsize = 20)
plt.ylabel("Y",fontsize = 20)

plt.show()
```

```python
#Histograms
x = np.random.normal(50,30,100000)

plt.figure(figsize = (20,8))
plt.hist(x,bins = 120,histtype = 'bar',edgecolor = 'red',color = 'black')
plt.xticks(range(-100,200,20))
plt.title("Histogram Plot",fontsize = 28)
plt.xlabel("X",fontsize = 20)
plt.ylabel("Y",fontsize = 20)
plt.show()
```

```python
#KDE (Kernel Density Estimation)
from scipy import stats

density = stats.kde.gaussian_kde(x)
plt.figure(figsize=(20,8))

X = np.linspace(min(x)-10, max(x)+10, 100)

plt.plot(X,density(X),color='#F325F2')
plt.fill_between(X,0,density(X),alpha=0.5, color='#FF7FF0')
plt.title("Kernel Density Estimation Plot",fontsize = 28)
plt.xlabel("X",fontsize = 20)
plt.ylabel("Y",fontsize = 20)

plt.show()
```

```python
#Combine Plots
values = np.random.normal(0,10,10000)
density = stats.kde.gaussian_kde(values)
x_range = np.linspace(min(values),max(values),10000)

plt.figure(figsize = (20,8))
plt.hist(values,bins = 120,histtype = 'bar',edgecolor = 'red',color = 'black',density = 1)
plt.plot(x_range,density(x_range),color = 'orange',linewidth = 5)
plt.title("Kernel Density Estimation Plot with Histogram Plot",fontsize = 28)
plt.xlabel("X",fontsize = 20)
plt.ylabel("Y",fontsize = 20)
plt.xticks(range(-40,50,10))

plt.show()
```

```python
#Bar plots
stars = [np.random.choice(range(1,6)) for counter in range(10)]
TV_Stars = [
    'Najwa',
    'Nawal',
    'Elissa',
    'Valeria',
    'Monatik',
    'Vera',
    'Antitila',
    'Kazka',
    'Maruv',
    'The Hardkiss'
]

plt.figure(figsize = (20,8))
plt.bar(TV_Stars,stars,edgecolor = 'yellow',linewidth = 7,color = 'blue')
plt.xticks(fontsize = 14)
plt.ylabel("Rating",fontsize = 22)
plt.xlabel("TV Star",fontsize = 22)
plt.grid(False)
plt.title("Bar Plot",fontsize=28)

plt.show()
```

```python
#Bar plots
rating_2020 = [np.random.choice(range(1,6)) for counter in range(10)]
rating_2021 = [np.random.choice(range(1,6)) for counter in range(10)]
TV_Stars = [
    'Najwa',
    'Nawal',
    'Elissa',
    'Valeria',
    'Monatik',
    'Vera',
    'Antitila',
    'Kazka',
    'Maruv',
    'The Hardkiss'
]

plt.figure(figsize = (20,8))
plt.bar(TV_Stars,rating_2020,linewidth = 4,color = 'blue',label = "2020 Rating",edgecolor = 'orange')
plt.bar(TV_Stars,rating_2021,linewidth = 4,color = 'green',bottom = rating_2020,label = "2021 Ratings",edgecolor = 'orange')
plt.xticks(fontsize = 14)
plt.ylabel("Rating",fontsize = 22)
plt.xlabel("TV Star",fontsize = 22)
plt.grid(True,axis = 'y')
plt.title("Bar Plot",fontsize=28)
plt.legend(fontsize = 15)
plt.yticks(range(1,11))
for x , y1 , y2 in zip(range(len(TV_Stars)),rating_2020,rating_2021) :
    plt.text(x - 0.3,y1-0.8,"{} stars".format(y1),fontsize = 14,color = 'red')
    plt.text(x - 0.3,y1 + y2-0.8,"{} stars".format(y2),fontsize = 14,color = 'red')

plt.show()
```

```python
#Box plots --used in outlier detection
values = [np.random.randint(1,11) if np.random.randint(1,10) !=3 else np.random.randint(18,21) for counter in range(30)]

plt.figure(figsize = (20,8))
plt.hist(values,bins = range(1,22,1) , histtype = 'bar',edgecolor = 'red',color = 'black',align = 'mid')
plt.yticks(range(0,15,1))
plt.xticks(range(1,22,1))
plt.grid(True,axis = 'y')
plt.title("Hitsogram Plot",fontsize=28)

plt.show()
```

```python
plt.figure(figsize = (20,8))
plt.boxplot(values)
plt.title("BoxPlot",fontsize=28)
plt.yticks(range(0,22,1))
plt.grid(True,axis = 'y')

plt.show()
```
