<h3>Iris dataset TSNE visualizing</h3>

```python
#Importing the needed packages
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
import pandas as pd 
```

```python
#Loading the dataset and initializing the needed variables.
iris = load_iris()
x = iris.data
y = iris.target 
```

```python
#Initializing a TSNE object and fitting it with the data.
#n_components : Dimension of the needed space.
tsne = TSNE(n_components=2, verbose=1, random_state=123,n_iter = 1000)
components = tsne.fit_transform(x)
```

```python
#Visualizing the dataset after its dimensions are reduced.
df = pd.DataFrame()
labels = iris.target_names[y]
df["y"] = labels
df["comp1"] = components[:,0]
df["comp2"] = components[:,1]

style.use("ggplot")
plt.figure(figsize = (20,8))
plt.title("Iris dataset T-SNE Visualization",fontsize = 22)
plot1 = sns.scatterplot(x="comp1", y="comp2", hue=df.y.tolist(),palette=sns.color_palette("Set1", n_colors = 3),data=df,s=100)
plt.xlabel("First Component of TSNE Output",fontsize = 18)
plt.ylabel("Second Component of TSNE Output",fontsize = 18)
plt.legend(fontsize = 18)
plt.show()
```
